```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
inputs = np.asarray([3.1, -4.5, 0.8, -2.7, 4.9])
```

```python
sigmoid(inputs)
```

```python
tanh(inputs)
```

```python
relu(inputs)
```
